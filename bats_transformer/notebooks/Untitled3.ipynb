```python
import matplotlib.pyplot as plt
import os
import re
import numpy as np
```

```python
BASE_DIR = "/home/vdesai/bats_logs_new/logs/"
files_all = os.listdir(BASE_DIR)
```

```python
files_all
```

```python
def plot_trends(id_string):
    p = re.compile(f"^{id_string}_[\d]*.log$")
    files = [f for f in files_all if p.match(f)]

    comp_list = []
    loss_list = []

    for f in sorted(files):
        comps = float(f.split("_")[-1].split(".")[0].strip())
        lines = open(os.path.join(BASE_DIR, f)).readlines()
        for l in lines:
            if l.startswith("50th percentile: "):
                loss = float(l.split(":")[1].strip())
                comp_list.append(comps)
                loss_list.append(loss)
                break
    
    plt.plot(comp_list, loss_list, 'x')
    plt.grid()
    plt.xlabel(f"{id_string}")
    plt.ylabel("Loss")
    print("Mean loss: ", np.mean(loss_list))
```

```python
plot_trends("0_25_quartile_seed")
```
