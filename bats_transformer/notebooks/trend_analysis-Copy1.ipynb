```python
import matplotlib.pyplot as plt
import os
import re
import numpy as np
```

```python
BASE_DIR = "/home/vdesai/bats_logs_new/logs/"
files_all = os.listdir(BASE_DIR)
```

```python
files_all
```

```python
def plot_trends(id_string):
    p = re.compile(f"^{id_string}_[\\d]*.log$")
    files_all = os.listdir(BASE_DIR)
    files = [f for f in files_all if p.match(f)]

    comp_list = []
    loss_list = []

    for f in sorted(files):
        comps = float(f.split("_")[-1].split(".")[0].strip())
        lines = open(os.path.join(BASE_DIR, f)).readlines()
        for l in lines:
            if l.startswith("50th percentile: "):
                loss = float(l.split(":")[1].strip())
                comp_list.append(comps)
                loss_list.append(loss)
                break

    plt.style.use('seaborn-darkgrid')  # Applying a better theme
    
    plt.figure(figsize=(10, 6))

    c_list = np.array(comp_list)
    l_list = np.array(loss_list)
    
    args = np.argsort(c_list)
    
    plt.plot(c_list[args], l_list[args], 'x-', linewidth=2, markersize=8, markeredgewidth=2)

    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.minorticks_on()
    plt.xlabel(f"{id_string}", fontsize=14, fontweight='bold')
    plt.ylabel("Loss", fontsize=14, fontweight='bold')
    plt.title(f"Trends for {id_string}", fontsize=16, fontweight='bold')
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    
    mean_loss = np.mean(loss_list)
    print("Mean loss: ", mean_loss)
    plt.figtext(0.15, 0.85, f'Mean loss: {mean_loss:.4f}', fontsize=12, fontweight='bold')

    plt.show()
```

```python
plot_trends("0_25_quartile_seed")
```
